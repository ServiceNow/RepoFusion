#### Effect of #PPCs

```python
import os
import json
base_dir = '/repo_data/repo_preprocessed_data/hole_pp_mappings/'

configs = os.listdir(base_dir)
for config in configs:
    unique_pp = 0
    encountered_holes = {}
    lines = open(os.path.join(base_dir, config)).readlines()
    print(config, len(lines))
    for line in lines:
        entry = json.loads(line)
        hole_id = entry['id']
        if hole_id not in encountered_holes:
            unique_pp += len(set(entry['prompt_proposals']))
            encountered_holes[hole_id] = 1
    print(config, len(encountered_holes), unique_pp/10000)
```
