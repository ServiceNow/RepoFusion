#### Effect of passage modes as a bar plot

```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams.update({'font.size': 25})
fig = plt.figure(figsize=(16, 8))

# Generate some example data
x = np.arange(4)
y1 = np.array([66.53, 72.78, 73.6, 74.82])
y2 = np.array([71.92, 71.92, 74.66, 74.73])

# Create a figure and axis object
ax = fig.add_subplot(111)

# Plot the bar charts
ax.bar(x, y1, width=0.15, label='768,32', color = 'navajowhite', edgecolor='black', hatch='o')
ax.bar(x+0.16, y2, width=0.15, label='512,63', color = 'green', edgecolor='black', hatch='+')

# Set the x-axis labels
ax.set_xticks(x + 0.075)
ax.set_xticklabels(['T-Rand', 'T-Rank', 'NT-Rank', 'NT-Prior-Last'])

# Add legends
ax.legend(loc='upper right', frameon=False, fontsize='small', bbox_to_anchor=(0.875, 1.0))

# Show the plot
plt.ylabel('Success Rate (%)')
plt.show()
```

#### Effect of repo context type as a bar plot

```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams.update({'font.size': 25})
fig = plt.figure(figsize=(16, 8))

# Generate some example data
x = np.arange(3)
y2 = np.array([50.33, 64.19, 74.66])
y1 = np.array([50.3, 52.98, 71.92])

# Create a figure and axis object
ax = fig.add_subplot(111)

# Plot the bar charts
ax.bar(x, y1, width=0.15, label='T-Rank', color = 'peru', edgecolor='black', hatch='x')
ax.bar(x+0.16, y2, width=0.15, label='NT-Rank', color = 'cyan', edgecolor='black', hatch='\\')

# Set the x-axis labels
ax.set_xticks(x + 0.075)
ax.set_xticklabels(['Random-NN', 'BM25', 'PP'])

# Add legends
ax.legend(loc='upper right', frameon=False, fontsize='small', bbox_to_anchor=(0.8, 1.0))

# Show the plot
plt.ylabel('Success Rate (%)')
plt.show()
```
